```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 파일 리스트 가져오기
directory = 'C:/Users/user/Desktop/kim/수업/24-01/240109/data'
flist = os.listdir(directory)

# CSV 파일만 필터링
csv_list = [f for f in flist if f.endswith('.csv')]

# 각 파일별로 결과 출력 및 자료실별 도서 개수 파이 차트로 시각화
for csv in csv_list:
    try:
        # 파일 읽기
        df = pd.read_csv(os.path.join(directory, csv), encoding='cp949')

        # 자료실을 기준으로 그룹화
        grouped_df = df.groupby(['자료실'])

        # 각 자료실별로 개수 세기
        count_per_room = grouped_df.size()

        # 결과 출력
        print(f"파일: {csv}")
        print(count_per_room)

        # 파이 차트 그리기
        plt.figure(figsize=(8, 8))
        plt.pie(count_per_room, labels=count_per_room.index, autopct='%1.1f%%', startangle=180)
        plt.title('자료실별 도서 개수')
        plt.show()

        print("=" * 30 + "\n")

    except UnicodeDecodeError:
        print(f"파일 {csv}의 인코딩이 'cp949' 또는 'euc-kr'이 아닐 수 있습니다.\n")
```

## 금액 범위 정해서 나누고 각 범위별 갯수

```python
for csv in csv_list:
    df = pd.read_csv(os.path.join(directory, csv), encoding='cp949')

    # '가격' 열을 숫자로 변환
    df['가격'] = pd.to_numeric(df['가격'], errors='coerce')
        
    # 가격 기준으로 분류
    below_10000 = df[df['가격'] <= 10000]
    between_10000_and_20000 = df[(df['가격'] > 10000) & (df['가격'] < 20000)]
    above_20000 = df[df['가격'] >= 20000]
    
    #가격별 갯수
    counts = [len(below_10000), len(between_10000_and_20000), len(above_20000)]

    # 결과 출력
    print(f"파일: {csv}")
    print(f"가격이 10000원 이하인 도서:\n{below_10000}\n")
    print(f"가격이 10000원 초과 20000원 이하인 도서:\n{between_10000_and_20000}\n")
    print(f"가격이 20000원 초과인 도서:\n{above_20000}\n")
    
    # 막대그래프 그리기
    categories = ['10000원 이하', '10000원 초과 20000원 미만', '20000원 이상']
    plt.bar(categories, counts)
    plt.xlabel('가격')
    plt.ylabel('도서 수')
    plt.title('가격 범주별 도서 수')
    plt.show()
        
    print("="*10 + "\n")        
```

## 가격 분포 히스토그램

```python
directory = 'C:/Users/user/Desktop/kim/수업/24-01/240109/data'
flist = os.listdir(directory)  # 주어진 리스트

csv_list = [f for f in flist if f.endswith('.csv')]

for csv in csv_list:
    df = pd.read_csv(os.path.join(directory, csv), encoding='cp949')

    # '가격' 열을 숫자로 변환
    df['가격'] = pd.to_numeric(df['가격'], errors='coerce')
    
    print(f"파일: {csv}")
    
    # 가격 분포 시각화
    plt.figure(figsize=(10, 6))
    sns.histplot(df['가격'], bins=20, kde=True)
    plt.xlabel('도서 가격')
    plt.xlim(0,50000)
    plt.ylabel('도서 수')
    plt.title('도서 가격 분포')
    plt.show()

    print("=" * 10 + "\n")
```

## 가격분포 산점도

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 파일 리스트 가져오기
directory = 'C:/Users/user/Desktop/kim/수업/24-01/240109/data'
flist = os.listdir(directory)

# CSV 파일만 필터링
csv_list = [f for f in flist if f.endswith('.csv')]

# 각 파일별로 결과 출력 및 가격 분포 산점도로 시각화
for csv in csv_list:
    try:
        # 파일 읽기
        df = pd.read_csv(os.path.join(directory, csv), encoding='cp949')

        # '가격' 열을 숫자로 변환
        df['가격'] = pd.to_numeric(df['가격'], errors='coerce')

        # 결과 출력
        print(f"파일: {csv}")

        # 가격 분포 시각화 - 산점도
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=df.index, y=df['가격'])
        plt.xlabel('도서')
        plt.ylabel('도서 가격')
        plt.ylim(0,60000)
        plt.title('도서 가격 분포 (산점도)')
        plt.show()

        print("=" * 10 + "\n")

    except UnicodeDecodeError:
        print(f"파일 {csv}의 인코딩이 'cp949' 또는 'euc-kr'이 아닐 수 있습니다.\n")
```
